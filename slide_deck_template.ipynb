## Investigation Overview

In this investigation, I wanted to investegate the trip main features that could affect the trip duration. The main features are age, user type, and gender.

## Dataset Overview

There are 1863721 entries in Gobike data from January 2019 to March 2020 with 16 features these include information about trip duration including:(Trip Duration, StartDate and EndDate) and station information including : (Start Station ID,Name, Latitude and Longitude, End Station ID, Name, Latitude and Longitude), along with information about the member these includes member_birth_year, member_gender and bike_id, user_type and finally the rental_access_method.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('2018-fordgobike-tripdata_clean.csv')
```

## Distribution of User typesÂ¶

Customers with a monthly membership are called Subscribers and others are called Customers, we can see that there's 90% subscribers and 10% customers.

```python
subscribers = df.query('user_type == "Subscriber"').bike_id.count() / df.bike_id.count()
customers = df.query('user_type == "Customer"').bike_id.count() / df.bike_id.count()
labels = ['Customer', 'Subscriber']
sizes = [customers, subscribers]
colors = [sb.color_palette()[1], sb.color_palette()[2]]
explode = (0, 0.1)
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, explode=explode)
plt.title('Distribustion of Biking Trips by User Type')
```

## Distribution of Trip Duration
Trip duration is right skewed, with focus on 5-15 mins.

```python
base_color = sb.color_palette()[6]
bins = np.arange(0, 70, 1)
ticks = np.arange(0, 70, 5)
plt.hist(data=df, x='duration_min', bins =bins, color=base_color)
plt.xticks(ticks, ticks)
plt.xlabel('Trip Duration in mins')
plt.title('Distribustion of Duration of biking trips in mins')
sb.despine()
```

## Distribution of Users Age
The distribution of age is in ranges between 25 to 45 years old.

```python
base_color = sb.color_palette()[4]
bins = np.arange(20, df['member_age'].max() + 1, 1)
ticks = np.arange(20, df['member_age'].max() + 5, 5)
plt.hist(data=df, x='member_age', color=base_color, bins=bins)
plt.xlabel('Ages of members')
plt.ylabel('Count')
plt.xticks(ticks, ticks)
plt.gca().yaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
plt.title('Distribustion of member ages')
sb.despine()
```
